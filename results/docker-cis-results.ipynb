```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
initial_scores = [19, 19, 17, 19, 19, 19, 17]
final_scores = [31, 31, 29, 31, 31, 31, 29]

fig, ax = plt.subplots()
bar_width = 0.35
X = np.arange(7)

p1 = plt.bar(X, initial_scores, bar_width, color='red',
label='CIS-Docker Benchmark scores before applying security rules')

p2 = plt.bar(X + bar_width, final_scores, bar_width,
color='orange',
label='CIS-Docker Benchmark scores after applying security rules')

plt.xlabel('Containers', fontsize=14)
plt.ylabel('Scores', fontsize=14)
plt.xticks(X + (bar_width/2) , ("Nginx", "MongoDB", "Docker", "MySQL", "Node", "Python", "Postgres"))

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

ax.spines['left'].set_position(('outward', 10))
ax.spines['bottom'].set_position(('outward', 20))

plt.legend(loc="lower center", bbox_to_anchor=(0.5, -0.5))
plt.savefig('docker-cis_results.jpg', dpi=750, bbox_inches='tight')
```
